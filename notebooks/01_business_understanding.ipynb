# Spotify Popularity Prediction

# 1. Comprensión del Negocio

## 1.1 Descripción del proyecto
El presente proyecto tiene como objetivo desarrollar un modelo capaz de predecir la popularidad de canciones en Spotify a partir de sus características musicales. Para ello, se empleará un conjunto de datos obtenido de Kaggle y se seguirá la metodología CRISP-DM.

Como complemento al modelo predictivo, se realizará un análisis mediante técnicas de aprendizaje no supervisado para identificar grupos de canciones con características similares. Finalmente, se desarrollará una aplicación interactiva utilizando Shiny para Python, en la cual se visualizarán los resultados obtenidos.

## 1.2 Planteamiento del problema
Actualmente existen millones de canciones disponibles en plataformas de streaming, lo que dificulta comprende cuáles características musicales se relacionan con su popularidad. Aunque existen factores externos que también influyen en el éxito de una canción, resulta interesante analizar hasta qué punto sus caraterísticas musicales permiten estimar su nivel de popularidad.

## 1.3 Objetivos del proyecto
**Objetivo general**

Desarrollar un modelo de aprendizaje automático capaz de predecir la popularidad de canciones en Spotify a partir de sus características musicales.

**Objetivos específicos**
* Analizar las características del conjunto de datos mediante técnicas de análisis exploratorio para comprenter su estructura y calidad
* Desarrollar y comparar diferentes modelos de regresión para identificar el modelo con mejor capacidad predictiva de la popularidad de las canciones.
* Indetificar grupos de canciones con características musicales similares mediante técnicas de aprendizaje no supervisado e integrar los resultados en una aplicación interactiva desarrollada con Shiny para Python.

## 1.4 Criterios de éxito
El proyecto se considerará exitoso si cumple con los siguientes criterios:

* Se desarrolla un modelo de regresión con capacidad predictiva adecuada evaluado mediante métricas como MAE, RMSE y R^2
* Se comparan distintos modelos de aprendizaje automático y se justigica la selección del modelo final con base en su desempeño.
* Se identifican grupos de canciones con características musicales similares que aporten información complementaria al análisis.
* Se implementa una aplicación funcional en Shiny para Python que permita utilizar el modelo de predicción y visualizar los resultados obtenidos.

## 1.5 Alcance del proyecto
El presente proyecto contempla el desarrollo de un modelo de regresión para estimar la popularidad de canciones utilizando únicamente sus característcas musicales. Asimismo, incluye el análisis exploratorio del conjunto de datos, la comparación de distintos modelos de aprendizaje automático, la aplicación de técnicas de clustering para identificar grupos de canciones con características similares y el desarrollo de una aplicación interactiva en Shiny para Python.

El proyecto no contempla el análisis de letras de canciones, el procesamiento de audio, la construcción de sistemas de recomendación musical ni la predicción de tendencias temporales o del éxito comercial de artistas.

## 1.6 Hipótesis inicial
Se plantea la hipótesis de que es posible estimar, por lo menos parcialmente, la popularidad de una canción utilizando únicamente sus características musicales. Asimismo, se espera que algunos atributos musicales presente una mayor influencia que otros en la capacidad predictiva del modelo y que existan grupos de canciones con características similares que compartan comportamientos semejantes en términos de popularidad.

## 1.7 Metodología propuesta
El proyecto seguirá la metodología CRISP-DM, debido a que proporciona una estructura organizada y ampliamente utilizada para el desarrollo de proyectos de minería de datos y aprendizaje automático.

Las etapas de la metodología se aplicarán de la siguiente manera:
* Business understanding: Definir el problema, los objetivos y el alcance del proyecto.
* Data Understanding: Explorar el conjunto de datos para comprender sus características, distribución y calidad.
* Data preparation: Realizar el prepocesamiento de los datos, incluyendo limpieza, transformación y selección de variables.
* Modeling: Desarollar y comparar distintos modelos de regresión, así como aplicar técnicas de clustering para el análisis complementario de los datos.
* Evaluation: Evaluar el desempeño de los modelos utilizando métricas apropiadas y analizar la utilidad de los resultados obtenidos.
* Deployment: Implementar una aplicación interactiva en Shiny que permita utilizar el modelo desarrollado y visualizar Los resultados del análisis.
